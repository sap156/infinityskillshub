## Verify Everything Works

**Make sure you've:**
1.  Set up your OpenAI API key in Colab Secrets (key icon in sidebar)
2.  Named the secret "OPENAI_API_KEY"
3.  Enabled "Notebook access"

**Then run the code below to test your connection:**

````python
# Import required libraries
from openai import OpenAI
from google.colab import userdata

api_key = userdata.get('OPENAI_API_KEY')

# Initialize OpenAI client
client = OpenAI(api_key=api_key)

# Set up secure connection
# Create OpenAI client
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
         "role": "user",
         "content": "Say Hello to our Data Professionals who are starting their AI journey!!"
         }
    ]
)

# Send API request


# Print response
print(response)
print("\n Connection Successful!!")
````

In our series, we will frequently work with JSON objects because they are easy to parse and manipulate in Python. Since the responses are structured as dictionaries, it allows us to efficiently extract and transform the data into a format suitable for various data engineering workflows.

**Why JSON for Data Work?**
- Structured and predictable
- Easy to parse and manipulate

````python
"""
Test AI's ability to return structured JSON data
"""
response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{
                "role": "user",
                "content": "Give me 3 tips for working with data. Return as JSON format with 'tips' array containing objects with 'title' and 'description' fields."
            }],
            response_format={"type": "json_object"}
        )

print("Raw Output")
print(response)

print("\n\nJSON Output")


print("\n\nFormatted JSON Output")


````
